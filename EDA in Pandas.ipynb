# Notable observations

-The dataset covers a span from 1970 to 2022, providing a comprehensive view of population changes over several decades.

-There is a noticeable increase in population over the years, with the mean population rising from around 16 million in 1970 to over 34 million in 2022.

-The dataset contains a wide range of population densities, with a minimum of 0.03 people per square kilometer and a maximum of 23172.27 people per square kilometer.

-The growth rate ranges from a minimum of 0.91% to a maximum of 1.07%, suggesting relatively consistent growth patterns.

-The maximum population value is exceptionally high (e.g., 1.43 billion in 2022), which may indicate the presence of outliers or a highly populous country in the dataset.
Similarly, the maximum world population percentage is 17.88%, which is notably higher than the mean.

-The dataset shows a right-skewed distribution of population sizes, with a few countries having significantly larger populations compared to the majority.

-The average land area is approximately 581,664 square kilometers, with a wide range of values.

-The minimum area is 1 square kilometer, which may suggest data entry issues or anomalies.

```python
sns.heatmap(df.corr(), annot = True)

plt.rcParams['figure.figsize'] = (20,7)

plt.show()
```

```python
df.groupby('Continent').mean().sort_values(by="2022 Population",ascending=False)
```
